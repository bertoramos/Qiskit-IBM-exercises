# Quantum Computing Programming : Qiskit

```python
from qiskit import *
from math import pi
from qiskit.visualization import plot_bloch_multivector
```

## Gates

### Pauli Gates: X, Y & Z Gate

```python
qc = QuantumCircuit(3)
qc.x(0)
qc.y(1)
qc.z(2)
qc.draw('mpl')
```

```python
backend = Aer.get_backend('statevector_simulator')
out = execute(qc, backend).result().get_statevector()
plot_bloch_multivector(out)
```

### Hadamard Gate

```python
qc1 = QuantumCircuit(1)
qc1.h(0)
qc1.draw('mpl')
```

```python
backend = Aer.get_backend('statevector_simulator')
out = execute(qc1, backend).result().get_statevector()
plot_bloch_multivector(out)
```

Confirm that by applying the H-gate twice in succession, the final state is always the same as the initial state

```python
qc2 = QuantumCircuit(1)
qc2.x(0)
qc2.h(0)
qc2.h(0)
qc2.draw('mpl')
```

```python
backend = Aer.get_backend('statevector_simulator')
out = execute(qc2, backend).result().get_statevector()
plot_bloch_multivector(out)
```
